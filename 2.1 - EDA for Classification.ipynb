## Feature influence on target

Although the graphs already show which characteristics influence the Health Status, I am going to try to identify, using the Anova table and the Tukey contrast, which independent features have the most impact on the Health Status since it is categorical.

### Auxiliary function

```python
# Function: Anova & possible Tukey
def anova_tukey(feature):
    # Calculate ANOVA
    model = ols(f'{feature} ~ C(Health_Status)', data=data).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    print(anova_table)

    # If ANOVA is significant, do the Tukey Test
    if anova_table['PR(>F)'].iloc[0] < 0.05:
        tukey = pairwise_tukeyhsd(endog=data[feature], groups=data['Health_Status'], alpha=0.05)
        print(tukey)
        tukey.plot_simultaneous()
        plt.title(f'Tukey Test Results for {feature}')
        plt.show()
```
